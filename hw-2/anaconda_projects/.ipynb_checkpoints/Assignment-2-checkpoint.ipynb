### 1. K-Nearest-Neighbor (KNN) classification on Newsgroups

#### For this problem you will use a subset of the 20 Newsgroup data set. The full data set contains 20,000 newsgroup documents, partitioned (nearly) evenly across 20 different newsgroups and has been often used for experiments in text applications of machine learning techniques, such as text classification and text clustering. The subset for this assignment includes 2,500 documents (newsgroup posts), each belonging to one of 5 categories windows (0), crypt (1), christian (2), hockey (3), forsale (4). The documents are represented by 9328 terms (stems). The dictionary (vocabulary) for the data set is given in the file "newsgroups5-terms.txt". The data has already been split into a 80% training set (newsgroups5-train.csv) and a 20% test set (newsgroups5-test.csv). The class labels for the training and test data are also provided in separate "label" files (with each row corresponding to the label for each test or training document). The training and test data are in term by document format, containing a row for each term in the vocabulary and a column for each document. The values in the table represent raw term occurrence counts for each document.
