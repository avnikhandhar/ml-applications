```python
# function to visualize all parameters as per its data type
def visualize_attributes(df):
 for col in df.columns:
    plt.figure(figsize=(6, 4))
    if df[col].dtype in ['int64', 'float64']:
        df[col].plot(kind='hist')
        plt.title(f'Histogram of {col}')
    else: # Assuming categorical
        df[col].value_counts().plot(kind='bar')
        plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count' if df[col].dtype == 'object' else 'Frequency')
    plt.tight_layout()
    plt.show()
```

```python
visualize_attributes(private_bachelors)
```

```python
# funnction to plot the correlations
def plot_correlations(data,title):
    plt.figure(figsize=(8,4))
    data.plot(kind='bar')
    plt.title(title)
    plt.ylabel('Correlation values')
    plt.axhline(0, color='black', linewidth=0.8)
    plt.show()
```

### 11. Use Min-Max Normalization to transform the values of the attribute hours-per-week the range 0.0-1.0 (without changing the original data). Next, perform zscore normalization (on the original data) to standardize the values of all numeric attributes (age, hours-per-week, education-num). The latter step should be performed on all three attributes at the same time instead of one-by-one (you may wish to first create a separate dataframe with only these attributes and perform the operation on the whole dataframe. Note: for this problem, you should write your own code to perform the normalization; do not use pre-existing functions such as scikit-learn's MinMaxScaler(). Finally, show the top 10 rows of the three versions of the hours-per-week attribute (original, normalized, and standardized) side-by-side in a new dataframe.

```python
# Min-Max Normalization for hours-per-week

# min hours value
min_hour = adult['hours-per-week'].min()

# max hours value
max_hour = adult['hours-per-week'].max()

normalized_hours = (adult['hours-per-week'] - min_hour) / (max_hour - min_hour)

normalized_hours.head(10)
```

```python
# dataframe with only the numeric attributes
adult_numeric = adult[["age","hours-per-week","education-num"]]

# standardization
adult_standardized = (adult_numeric - adult_numeric.mean()) / adult_numeric.std()

adult_standardized.head()
```

```python
# comparing the three varsions of hours-per-week
compare_hours = pd.DataFrame({
    'original hours': adult['hours-per-week'],
    'normalized hours': normalized_hours,
    'standardized hours': adult_standardized['hours-per-week']
})
```

```python
# viewing first 10 rows for comparison
compare_hours.head(10)
```
